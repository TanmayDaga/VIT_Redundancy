```python
!pip install timm
from timm import create_model

PATH_OF_IMAGE:str = "./img.jpeg"
MODEL_NAME:str = "vit_base_patch16_224"

model = create_model(MODEL_NAME, pretrained=True)
model.eval()
```

## Preprocessor

```python
from PIL import Image
from timm.data import resolve_data_config
from torchvision.transforms import Compose
from timm.data.transforms_factory import create_transform
from torch import Tensor

img = Image.open(PATH_OF_IMAGE).convert("RGB")
config:dict = resolve_data_config({}, model=MODEL_NAME)
transformation_compose:Compose = create_transform(**config)
input_tensor:Tensor = transformation_compose(img).unsqueeze(0)
input_tensor[0][0]

```

```python
import torch.nn.functional as F

layer = model.patch_embed.proj

input_unfolded = F.unfold(input_tensor, kernel_size=(16,16), stride = (16,16))
weight_matrix = layer.weight.view(768, -1)



print(input_unfolded.shape)
print(weight_matrix.dtype)
```

```python
import multiplier
multiplier.init()
```

## Patch Embedding

```python
result = multiplier.mat_mul(weight_matrix.tolist(), input_unfolded[0].tolist(), "Conv2d", MODEL_NAME)
```
